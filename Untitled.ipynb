```python
name = 'Cardinalidae'
filt = 'SA'
taxon = df[(df['filter'] == filt) & (df['name'] == name)]

fig, axs = plt.subplots(3,4, figsize = (10,10))
for idx, i in enumerate(taxon.index):
    row = idx // 4
    col = idx % 4
    ax = axs[row,col]
    preds = [int(n) for n in taxon.loc[i,'preds'][1:-1].split(',')]
    ax.hist(preds, bins = 30)
    ax.vlines(taxon.loc[i, 'sub_taxa'], *ax.get_ylim(), color = 'red')
    ax.set_title(round(taxon.loc[i, 'thresh'],2))
fig.suptitle(name)
plt.tight_layout()
```
